```python
columns_to_remove = ['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber']
hr_data_cleaned = hr_data.drop(columns=columns_to_remove)
hr_data_cleaned.head()
```

```python
columns_rename = {
    'Attrition': 'EmployeeAttrition',
    'BusinessTravel': 'BusinessTravelFrequency',
    'DailyRate': 'DailySalaryRate',
    'DistanceFromHome': 'CommuteDistance',
    'Education': 'EducationLevel',
    'EducationField': 'FieldOfEducation',
    'JobInvolvement': 'JobEngagement',
    'JobSatisfaction': 'JobHappiness',
    'NumCompaniesWorked': 'CompaniesWorkedAt',
    'PercentSalaryHike': 'SalaryIncreasePercent',
    'PerformanceRating': 'JobPerformanceRating',
    'RelationshipSatisfaction': 'InterpersonalSatisfaction',
    'StockOptionLevel': 'StockOptions',
    'TotalWorkingYears': 'TotalYearsWorked',
    'TrainingTimesLastYear': 'TrainingLastYear',
    'WorkLifeBalance': 'WorkLifeBalanceRating',
    'YearsAtCompany': 'YearsWithCompany',
    'YearsInCurrentRole': 'YearsInRole',
    'YearsSinceLastPromotion': 'YearsSincePromotion',
    'YearsWithCurrManager': 'YearsWithManager'
}
hr_data_cleaned.rename(columns=columns_rename, inplace=True)
hr_data_cleaned.head()
```

```python
hr_data_cleaned.drop_duplicates(inplace=True)
hr_data_cleaned.head()
```

```python
nan_values = hr_data_cleaned.isna().sum()
nan_values
```

```python
categoried_columns = ['EmployeeAttrition', 'BusinessTravelFrequency', 'Department', 'FieldOfEducation', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']
for column in categoried_columns:
    hr_data_cleaned[column] = hr_data_cleaned[column].str.title()

unique_values = {column: hr_data_cleaned[column].unique() for column in categoried_columns}
unique_values
```
